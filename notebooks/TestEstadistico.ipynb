```python
import numpy as np


def mean_reciprocal_rank(rs):
    rs = (np.asarray(r).nonzero()[0] for r in rs)
    return np.mean([1. / (r[0] + 1) if r.size else 0. for r in rs])


def r_precision(r):
    r = np.asarray(r) != 0
    z = r.nonzero()[0]
    if not z.size:
        return 0.
    return np.mean(r[:z[-1] + 1])


def precision_at_k(r, k):
    assert k >= 1
    r = np.asarray(r)[:k] != 0
    if r.size != k:
        raise ValueError('Relevance score length < k')
    return np.mean(r)


def average_precision(r,max_k):
    r = np.asarray(r) != 0
    out = [precision_at_k(r, k + 1) for k in range(max_k) if r[k]]
    if not out:
        return 0.
    return np.mean(out)


def mean_average_precision_at_k(rs,max_k):
    return np.mean([average_precision(r,max_k) for r in rs])

# def dcg_at_k(r, k, method=0):
#     r = np.asfarray(r)[:k]
#     if r.size:
#         if method == 0:
#             return r[0] + np.sum(r[1:] / np.log2(np.arange(2, r.size + 1)))
#         elif method == 1:
#             return np.sum(r / np.log2(np.arange(2, r.size + 2)))
#         else:
#             raise ValueError('method must be 0 or 1.')
#     return 0.


# def ndcg_at_k(r, k, method=0):
#     dcg_max = dcg_at_k(sorted(r, reverse=True), k, method)
#     if not dcg_max:
#         return 0.
#     return dcg_at_k(r, k, method) / dcg_max

#KAGGLE
def dcg_at_k(r, k):
    r = np.asfarray(r)[:k]
    if r.size:
        return np.sum(np.subtract(np.power(2, r), 1) / np.log2(np.arange(2, r.size + 2)))
    return 0.


def ndcg_at_k(r, k):
    idcg = dcg_at_k(sorted(r, reverse=True), k)
    if not idcg:
        return 0.
    return dcg_at_k(r, k) / idcg
```

```python
#df = pd.DataFrame([list(s1), list(s2)],  columns =  ["A", "B", "C"])

max_k = 21
resultall = []
result_precision = []
result_hits = []
result_recall = []
result_fmeasure = []
result_ndcg_k = []
result_mean_average_precision = []

#Searchers Used
#print(searchers)
arr_searchers = pd.Series(['npmjs.com_optimal', 'npmsearch.com', 'M1', 'M2','M3', 'M4', 'BordaFuse', 'Cordorcet',
                           'Filter_google.com_hyp_match', 'Filter_bing.com_hyp_match'])
#Queries Used
#print(df['query'].tolist())
arr_queries = pd.Series(['download free music', 'detect text language', 'download web videos', 'check valid email address', 
                         'send sms', 'health tracker', 'quick sort algorithm', 'credit card validation', 
                         'third party authentication', 'calculate word similarity', 'mobilie app framework', 
                         'convert typewritten image to text', 'user authentication', 'filter adult content images', 
                         'lightweight 3D graphic library', 'convert text to speech', 'extract barcode from image', 
                         'mathematical functions', 'sentiment analysis', 'rank aggregation algorithms', 
                         'convert data formats', 'translate english to spanish', 'captcha authentication', 
                         'DOM manipulation utils', 'scraper'])


#Results by queries
dataframe_hits_by_k = []
dataframe_precision_by_k = []
dataframe_recall_by_k = []
dataframe_fmeasure_by_k = []
dataframe_ndcg_by_k = []
dataframe_map_by_k = []

# Hits at k
for k in range(1,max_k):
    for query in range(arr_queries.size):
        searchersrow = []
        row_hits = []

        for searcher in range(arr_searchers.size):
            sum_hits = 0

            idxq = df.index[df['query'] == arr_queries[query]].tolist()
            idxs = searchers.index(arr_searchers[searcher])
            sxq = df.iloc[idxq[0]][idxs+1]
            sxq = sxq.set_index('k')
            sum_hits += sxq.loc[k]['#hits']
            
            searcherprom = pd.DataFrame([[sum_hits]],  columns =  ["hits"])
            searchersrow.append(searcherprom)
            row_hits.append(sum_hits)

        resultall.append([arr_queries[query]] + searchersrow)
        result_hits.append([arr_queries[query]] + row_hits)

    columnsk = ['query'] + arr_searchers.tolist()
    dataframek = pd.DataFrame(resultall, columns =  columnsk)
    dataframek = dataframek.set_index('query')
    dataframe_hits = pd.DataFrame(result_hits, columns =  columnsk)
    dataframe_hits_by_k.append(dataframe_hits)

# Precision at k
for k in range(1,max_k):
    for query in range(arr_queries.size):
        searchersrow = []
        row_precision = []

        for searcher in range(arr_searchers.size):
            sum_precision = 0

            #print(arr_queries[query])
            idxq = df.index[df['query'] == arr_queries[query]].tolist()
            idxs = searchers.index(arr_searchers[searcher])
            sxq = df.iloc[idxq[0]][idxs+1]
            sxq = sxq.set_index('k')
            sum_precision += sxq.loc[k]['precision']

            searcherprom = pd.DataFrame([[sum_precision]],  columns =  ["precision"])
            searchersrow.append(searcherprom)
            row_precision.append(sum_precision)

        resultall.append([arr_queries[query]] + searchersrow)
        result_precision.append([arr_queries[query]] +row_precision)
        
    columnsk = ['query'] + arr_searchers.tolist()
    dataframek = pd.DataFrame(resultall, columns =  columnsk)
    dataframek = dataframek.set_index('query')
    dataframe_precision = pd.DataFrame(result_precision, columns =  columnsk)
    dataframe_precision_by_k.append(dataframe_precision)

# Recall at k
for k in range(1,max_k):
    for query in range(arr_queries.size):
        searchersrow = []
        row_recall = []

        for searcher in range(arr_searchers.size):
            sum_recall = 0

            idxq = df.index[df['query'] == arr_queries[query]].tolist()
            idxs = searchers.index(arr_searchers[searcher])
            sxq = df.iloc[idxq[0]][idxs+1]
            sxq = sxq.set_index('k')
            sum_recall += sxq.loc[k]['recall']

            searcherprom = pd.DataFrame([[sum_recall]],  columns =  ["ncdg"])
            searchersrow.append(searcherprom)
            row_recall.append(sum_recall)

        resultall.append([arr_queries[query]] + searchersrow)
        result_recall.append([arr_queries[query]]+row_recall)

    columnsk = ['query'] + arr_searchers.tolist()
    dataframek = pd.DataFrame(resultall, columns =  columnsk)
    dataframek = dataframek.set_index('query')
    dataframe_recall = pd.DataFrame(result_recall, columns =  columnsk)
    dataframe_recall_by_k.append(dataframe_recall)    

# F-measure at k
for k in range(1,max_k):
    for query in range(arr_queries.size):
        searchersrow = []
        row_fmeasure = []

        for searcher in range(arr_searchers.size):
            sum_fmeasure = 0

            idxq = df.index[df['query'] == arr_queries[query]].tolist()
            idxs = searchers.index(arr_searchers[searcher])
            sxq = df.iloc[idxq[0]][idxs+1]
            sxq = sxq.set_index('k')
            sum_fmeasure += sxq.loc[k]['fmeasure']
            prom_fmeasure = sum_fmeasure

            searcherprom = pd.DataFrame([[prom_fmeasure]],  columns =  ["ncdg"])
            searchersrow.append(searcherprom)
            row_fmeasure.append(prom_fmeasure)

        resultall.append([arr_queries[query]] + searchersrow)
        result_fmeasure.append([arr_queries[query]]+row_fmeasure)

    columnsk = ['query'] + arr_searchers.tolist()
    dataframek = pd.DataFrame(resultall, columns =  columnsk)
    dataframek = dataframek.set_index('query')
    dataframe_fmeasure = pd.DataFrame(result_fmeasure, columns =  columnsk)

    dataframe_fmeasure_by_k.append(dataframe_fmeasure)
    
# NDCG at k
for k in range(1,max_k):
    for query in range(arr_queries.size):
        searchersrow = []
        row_ndcg_k = []

        for searcher in range(arr_searchers.size):
            sum_ndcg_k = 0

            idxq = df.index[df['query'] == arr_queries[query]].tolist()
            idxs = searchers.index(arr_searchers[searcher])
            sxq = df.iloc[idxq[0]][idxs+1]
            sxq = sxq.set_index('k')
            sum_ndcg_k += ndcg_at_k(sxq['hit'],k)
            prom_ndcg_k = sum_ndcg_k

            searcherprom = pd.DataFrame([[prom_ndcg_k]],  columns =  ["ncdg"])
            searchersrow.append(searcherprom)
            row_ndcg_k.append(prom_ndcg_k)

        result_ndcg_k.append([arr_queries[query]] + row_ndcg_k)

    columnsk = ['query'] + arr_searchers.tolist()
    dataframek = pd.DataFrame(resultall, columns =  columnsk)
    dataframek = dataframek.set_index('query')
    dataframek.head(3)

    dataframe_ndcg_k = pd.DataFrame(result_ndcg_k, columns =  columnsk)
    dataframe_ndcg_by_k.append(dataframe_ndcg_k)

# MAP at k

for k in range(1,max_k):
    for query in range(arr_queries.size):
        searchersrow = []
        row_mean_average_precision = []

        for searcher in range(arr_searchers.size):
            arr_rs = []    

            #print(arr_queries[query])
            idxq = df.index[df['query'] == arr_queries[query]].tolist()
            idxs = searchers.index(arr_searchers[searcher])
            sxq = df.iloc[idxq[0]][idxs+1]
            sxq = sxq.set_index('k')
            arr_rs.append(sxq['hit']);
            mean_average_precision = mean_average_precision_at_k(arr_rs,k)

            searcherprom = pd.DataFrame([[mean_average_precision]],  columns =  ["ncdg"])
            searchersrow.append(searcherprom)
            row_mean_average_precision.append(mean_average_precision)

        result_mean_average_precision.append([arr_queries[query]]+row_mean_average_precision)

    columnsk = ['query'] + arr_searchers.tolist()
    dataframek = pd.DataFrame(resultall, columns =  columnsk)
    dataframek = dataframek.set_index('query')
    
    dataframe_average_precision = pd.DataFrame(result_mean_average_precision, columns =  columnsk)
    dataframe_map_by_k.append(dataframe_average_precision)

#dataframe_ndcg_k.to_csv('img/ndcg10ByQuery.csv')
```
